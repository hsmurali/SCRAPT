```python
from os import listdir, mkdir
from os.path import isdir, isfile

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from Cluster_Evaluation_Utils import *
import matplotlib.ticker as mtick

from matplotlib.ticker import FuncFormatter
import pickle

f = mtick.ScalarFormatter(useOffset=False, useMathText=True)
g = lambda x,pos : "${}$".format(f._formatSciNotation('%1.1e' % x))


rcParams = {'font.size': 30 , 'font.weight': 'normal', 'font.family': 'sans-serif',
            'axes.unicode_minus':False, 'axes.labelweight':'normal'}
plt.rcParams.update(rcParams)
```

```python
data_dir = '/Users/harihara/Research-Activities/Data/SCRAPT/Lupus-Microbiome-Published/'
outdir = '/Users/harihara/Research-Activities/Plots/Iterative_Clustering/Lupus-Microbiome-MT/'
n = 950180
```

```python
plt.rcParams.update(rcParams)
d_95 = pickle.load(open(data_dir+"sim_0.95.pkl","rb"))
y_scrapt, x_scrapt = d_95[0.5]['Adaptive_With_Modeshifting']['Fragmentation_Measure']
df_meshclust_95 = pd.read_csv(data_dir+'output_meshclust_3.0_0.95.txt', sep = "\t", 
                              names = ['Cluster','Sequence', 'Identity','Category'])
df_meshclust_95_cm = df_meshclust_95[(df_meshclust_95['Category'] == 'C') | 
                                     (df_meshclust_95['Category'] == 'M')]
df_meshclust_95_cme = df_meshclust_95[(df_meshclust_95['Category'] == 'C') | 
                                      (df_meshclust_95['Category'] == 'M') |
                                      (df_meshclust_95['Category'] == 'E')]
df_meshclust_95_cme = df_meshclust_95_cme.groupby('Cluster').count()[['Sequence']]
y_meshclust_cme, x_meshclust_cme = Compute_Fragmentation_Measure(df_meshclust_95_cme['Sequence'].tolist())

df_meshclust_95_cm = df_meshclust_95_cm.groupby('Cluster').count()[['Sequence']]
y_meshclust_cm, x_meshclust_cm = Compute_Fragmentation_Measure(df_meshclust_95_cm['Sequence'].tolist() + 
                                                               [1]*(n-sum(df_meshclust_95_cm['Sequence'].tolist())))

fig,ax = plt.subplots(1,1,figsize = (16,10))
ax.plot(np.array(x_scrapt)/n*100.0,y_scrapt, linewidth=5, label = 'SCRAPT')
ax.plot(np.array(x_meshclust_cm)/n*100.0,y_meshclust_cm, linewidth=5, label = 'MeShClust 3.0')
ax.plot(np.array(x_meshclust_cme)/n*100.0,y_meshclust_cme, linewidth=5, label = 'MeShClust 3.0(Extended Members)',
        color = 'grey')

ax.set_xlim(ax.get_xlim()[::-1])
ylim = ax.get_ylim()
ax.yaxis.set_major_formatter(mtick.FuncFormatter(g))
ax.legend(loc=2, frameon = False)
ax.set_ylabel('Total number of sequences from \n'+r'clusters of size $\geq$ x')
ax.set_xlabel('x: Cluster size (% total number of sequences, N = '+str(n)+
              ')\n Ordered from the largest to the smallest')
fig.tight_layout()
fig.savefig(outdir+'MESHCLUST_95.pdf')
```
